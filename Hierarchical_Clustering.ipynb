```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
data = pd.read_csv('Mall.csv') 
```

```python
data.head()
```

```python
X = data.iloc[:,[3,4]] # Splitting X and y
```

```python
# Using dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
```

```python
# Using method= "ward" (minimizes the variance of the clusters being merged)
plt.figure(figsize=[20,10])
dendrogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Using method= "average" (uses the average of the distances of each observation of two datasets)
plt.figure(figsize=[20,10])
dendrogram = sch.dendrogram(sch.linkage(X,method = 'average'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Using method= "complete" (uses the maximum distances between all observations of the two datasets)
plt.figure(figsize=[20,10])
dendrogram = sch.dendrogram(sch.linkage(X,method = 'complete'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Fitting hierarchical clustering to the dataset
# Visualizing the clusters
# only for plotting the clusters in 2 dimensions
# linkage = "ward"
hc = AgglomerativeClustering(n_clusters = 5,
                             affinity = 'euclidean',
                             linkage = 'ward')
y_hc = hc.fit_predict(X)
plt.figure(figsize=(20,10))
plt.scatter(X.values[y_hc == 0 , 0], X.values[y_hc == 0, 1], s = 100, c= 'red', label = 'Careful')
plt.scatter(X.values[y_hc == 1 , 0], X.values[y_hc == 1, 1], s = 100, c= 'blue', label = 'Standard')
plt.scatter(X.values[y_hc == 2 , 0], X.values[y_hc == 2, 1], s = 100, c= 'green', label = 'Target')
plt.scatter(X.values[y_hc == 3 , 0], X.values[y_hc == 3, 1], s = 100, c= 'magenta', label = 'Careless')
plt.scatter(X.values[y_hc == 4 , 0], X.values[y_hc == 4, 1], s = 100, c= 'cyan', label = 'Sensible')
plt.title('Clusters of Clients')
plt.xlabel('Annual Income k$')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
